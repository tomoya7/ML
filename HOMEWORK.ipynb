```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.linear_model import LinearRegression
```

## DATASET

```python
np.random.seed(7)
x=np.random.random(100)*10
X=x.reshape(-1,1)
y=np.sin(x)+np.random.normal(0,0.5,size=100)
plt.scatter(x,y)
```

## model

```python
def polynomial_regression(degree):
    return Pipeline([
        ('poly',PolynomialFeatures(degree=degree)),
        ('lin_reg',LinearRegression())
    ])
```

## var & bais

```python
var,biases=[],[]
min_degree,max_degree=1,7
for i in range(min_degree,max_degree):
    poly_reg=polynomial_regression(i)
    poly_reg.fit(X,y)
    y_predict=poly_reg.predict(X)
    var.append(np.var(y_predict))
    bias=np.mean(np.abs(y-y_predict))
    biases.append(bias)
```

```python
plt.figure()
plt.plot(np.arange(min_degree,max_degree),var,label='var')
plt.plot(np.arange(min_degree,max_degree),np.array(biases)**2,'r',label='bias')
plt.legend()
```
